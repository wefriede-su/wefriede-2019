```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

website = 'https://www.basketball-reference.com/contracts/players.html'
salaries = pd.read_html(website)
salaries[0].columns = [ 'Rank', 'Player', 'Team', '2018-19', '2019-20', '2020-21', '2021-22', '2022-23', '2023-24', 'Signed_Using', 'Guaranteed']
salaries_df = pd.DataFrame(salaries[0])

website = 'https://www.basketball-reference.com/leagues/NBA_2019_per_game.html'
stats = pd.read_html(website)
stats_df = pd.DataFrame(stats[0])

try:
    salary_unused = 140000000
    count = 0
    salary_used = 0
    luxury_tax = 0
    players = []
    print("Pick 12 players. Copy and paste the player name to ensure correct spelling.")
    print(salaries_df[['Player','Team','2018-19']])
    while True:
        if count == 12:
            print("Maximum players reached.")
            break
        elif salary_used < 110000000:
            print(salaries_df[['Player','Team','2018-19']].sample(15))
            player = input("Enter a player: ")
            if salaries_df.empty == True:
                print("That player does not exist. Please start over.")
                break
            else:
                print("That player exists!")
            print(salaries_df[['Player','Team','2018-19']][salaries_df['Player'] == player].head(1))
            salary = int(input("Enter that player's salary: "))
            players.append(player)
            count = count + 1
            salary_used = salary_used + salary
            salary_unused = salary_unused - salary
            print("You have used $%d and have $%d left" % (salary_used, salary_unused))
        elif 110000000 < salary_used < 140000000:
            print(salaries_df[['Player','Team','2018-19']].sample(15))
            player = input("Enter a player: ")
            if salaries_df.empty == True:
                print("That player does not exist. Please start over.")
                break
            else:
                print("That player exists!")
            print(salaries_df[['Player','Team','2018-19']][salaries_df['Player'] == player].head(1))
            salary = int(input("Enter that player's salary: "))
            players.append(player)
            count = count + 1
            luxury_tax = 1.5*salary
            salary_used = salary_used + salary + luxury_tax
            salary_unused = salary_unused - salary - luxury_tax
            print("You are in the luxury tax. You have used $%d and have $%d left" % (salary_used, salary_unused))
        elif salary_used > 140000000:
            print("You have gone over and have to start again.")
            players = []
            salary_unused = 140000000
            count = 0
            salary_used = 0
            luxury_tax = 0
            
except TypeError:
    print("That is not a contract value.")

        
print(players)
while True:
    player_search = input("Enter a player you selected or press 'quit': ")
    if player_search == 'quit':
        break
    else:
        print(stats_df[['Player','Tm','PTS','TRB','AST','STL','BLK','TOV','FT%','eFG%','3P%']][stats_df['Player'] == player_search])
```
